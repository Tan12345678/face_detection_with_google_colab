```
face_cascade=cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

```
faces=face_cascade.detectMultiScale(gray2,1.1,4)
for (x, y, w, h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),1)
cv2_imshow(img)
cv2.waitKey()
```
